Encode Categorical Columns

```python
encoders = {}

categorical_columns = [
    "Sex",
    "ChestPainType",
    "RestingECG",
    "ExerciseAngina",
    "ST_Slope"
]

for column in categorical_columns:

    encoder = LabelEncoder()

    df[column] = encoder.fit_transform(df[column])

    encoders[column] = encoder
```

```python
print(df.head())
```

Split Features & Target

```python
X = df.drop("HeartDisease", axis=1)
y = df["HeartDisease"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=33,stratify=y)
```

Feature Scaling

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Build Random Forest Model

```python
model = RandomForestClassifier(n_estimators=200,random_state=33)
```

Train Model

```python
model.fit(X_train_scaled, y_train)
```

Make Predictions

```python
y_pred = model.predict(X_test_scaled)
```

Calculate Evaluation Metrics

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

train_accuracy = model.score(X_train_scaled, y_train)
test_accuracy = model.score(X_test_scaled, y_test)
```

```python
print("\nTraining Accuracy : {:.2f}%".format(train_accuracy * 100))
print("Testing Accuracy  : {:.2f}%".format(test_accuracy * 100))

print("\nAccuracy Score : {:.4f}".format(accuracy))
print("Precision      : {:.4f}".format(precision))
print("Recall         : {:.4f}".format(recall))
print("F1 Score       : {:.4f}".format(f1))
```

Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

Classification Report

```python
print(classification_report(y_test, y_pred))
```

Feature Importance

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
})

importance = importance.sort_values(by="Importance",ascending=False)
```

```python
print(importance)
```

Save Trained Model

```python
joblib.dump(model, "heart_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(encoders, "label_encoders.pkl")
```
